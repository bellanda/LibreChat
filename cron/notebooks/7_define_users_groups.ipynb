```python
import json
import os

import dotenv
from pymongo import MongoClient

dotenv.load_dotenv(override=True)

LOCAL_MONGO_URI = os.getenv("MONGO_URI")
SERVER_MONGO_URI = os.getenv("MONGO_URI_SERVER")

local_client = MongoClient(LOCAL_MONGO_URI)
server_client = MongoClient(SERVER_MONGO_URI)

users_groups = {
    "gb810437": "ultimate",
}
```

```python
from pymongo import UpdateOne

operations = [UpdateOne({"username": username}, {"$set": {"group": group}}) for username, group in users_groups.items()]

if operations:
    result = server_client.LibreChat.users.bulk_write(operations)
    print(f"Documentos correspondentes: {result.matched_count}")
    print(f"Documentos modificados: {result.modified_count}")
```
